### Init

```python
import sys
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df_diff_data = pd.read_csv("energies_diff_4S0O_2K7W.csv", sep=";", index_col=0)
df_diff = pd.DataFrame(df_diff_data)
df_diff = df_diff.drop("Residue #", axis=1)
```

```python
fig, ax = plt.subplots(figsize=(10, 8))

sns.boxplot(df_diff)
plt.title("Difference in free energy across amino acids - 4S0O vs 2K7W")
```

#### Difference in free energy across domains
BH1: res 98-118\
BH2: res 150-165\
BH3: res 59-73

```python
### BH3 difference in free energy:
fig, ax = plt.subplots(figsize=(10, 8))
df_diff_BH3 = df_diff.iloc[59:74]
sns.boxplot(df_diff_BH3.T, ax=ax)
ax.set_xlabel("BH3 residues")
plt.title("BH3 difference in free energy - 4S0O vs 2K7W")
```

```python
### BH1 difference in free energy:
fig, ax = plt.subplots(figsize=(10, 8))
df_diff_BH3 = df_diff.iloc[98:119]
sns.boxplot(df_diff_BH3.T, ax=ax)
ax.set_xlabel("BH1 residues")
plt.title("BH1 difference in free energy - 4S0O vs 2K7W")
```

```python
### BH2 difference in free energy:
fig, ax = plt.subplots(figsize=(10, 8))
df_diff_BH3 = df_diff.iloc[150:166]
sns.boxplot(df_diff_BH3.T, ax=ax)
ax.set_xlabel("BH2 residues")
plt.title("BH2 difference in free energy - 4S0O vs 2K7W")
```
